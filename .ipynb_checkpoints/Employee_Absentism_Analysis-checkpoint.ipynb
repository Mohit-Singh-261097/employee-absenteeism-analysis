```python
# Frequency count of JobTitle and DepartmentName
job_department_count = df.groupby(['JobTitle', 'DepartmentName']).size().unstack(fill_value=0)

# Heatmap of JobTitle and DepartmentName
plt.figure(figsize=(12, 8))
sns.heatmap(job_department_count, cmap='Blues', annot=False)
plt.title('Job Title Distribution Across Departments')
plt.xlabel('Department')
plt.ylabel('Job Title')
plt.show()

# Chi-Square Test to check if Job Title distribution is independent of Department

# Create a contingency table
contingency_table = pd.crosstab(df['JobTitle'], df['DepartmentName'])

# Perform Chi-Square Test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f'Chi-Square Test p-value for JobTitle and Department: {p}')
```

```python
#Boxplot for Age, Gender, and Absent Hours


plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='AbsentHours', hue=pd.cut(df['Age'], bins=[20, 30, 40, 50, 60]), data=df)
plt.title('Absent Hours by Gender and Age Group')
plt.show()

df_sample = df.sample(frac = 0.2, axis = 0)
df_sample['AgeGroup'] = pd.cut(df_sample['Age'], bins=[20, 30, 40, 50, 60])

# Filter out groups with zero variance in AbsentHours
filtered_sample = df_sample.groupby(['Gender', 'AgeGroup']).filter(lambda x: x['AbsentHours'].var() > 0)

# Run ANOVA after filtering
anova_model = ols('AbsentHours ~ C(Gender) + C(AgeGroup)', data=filtered_sample).fit()
anova_table = sm.stats.anova_lm(anova_model, typ=2)
print(anova_table)
```

```python
# Check if there are any NaN or invalid values in AgeGroup or AbsentHours
print(df_sample[['AgeGroup', 'AbsentHours']].info())
print(df_sample[['AgeGroup', 'AbsentHours']].isna().sum())

# Drop rows with NaN values in AgeGroup or AbsentHours
df_sample_clean = df_sample.dropna(subset=['AgeGroup', 'AbsentHours'])

# Verify that AgeGroup has only valid categories (strings, not intervals)
print(df_sample_clean['AgeGroup'].unique())

# Now try running Tukey's HSD test again
tukey_age = pairwise_tukeyhsd(endog=df_sample_clean['AbsentHours'], groups=df_sample_clean['AgeGroup'], alpha=0.05)
print(tukey_age)

# Tukey's HSD test for Gender
tukey_gender = pairwise_tukeyhsd(endog=df_sample_clean['AbsentHours'], groups=df_sample_clean['Gender'], alpha=0.05)
print(tukey_gender)
```

```python
df['churn'] = ((df['LengthService']>2.0) & (df['AbsentHours'] > 40.0)).astype('int')
```

```python
y = df['churn'].values
x =df[['Gender', 'JobTitle','DepartmentName', 'Age', 'LengthService',
       'AbsentHours', 'BusinessUnit', 'JobTitle_BuisnessUnit']]
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
scaler.fit(x_train[numeric_features])
x_train.loc[:, numeric_features] = scaler.transform(x_train[numeric_features])
x_test.loc[:, numeric_features] = scaler.transform(x_test[numeric_features])
```

```python
x_train_encoded = pd.get_dummies(x_train, drop_first=True)
x_test_encoded = pd.get_dummies(x_test, drop_first=True)

x_train_encoded.columns = x_train_encoded.columns.str.replace(r'[^\w\s]', '', regex=True)
x_test_encoded.columns = x_test_encoded.columns.str.replace(r'[^\w\s]', '', regex=True)

# Ensure consistency of feature names between train and test sets


# Align the columns of the test set to match the training set
x_test_encoded = x_test_encoded.reindex(columns=x_train_encoded.columns, fill_value=0)
```

```python
x_train_encoded.head()
```

```python
x_test_encoded.head()
```

```python


# Define models and their hyperparameter grids
models = {
    'RandomForest': RandomForestClassifier(),
    'XGBoost': XGBClassifier(),
    'LogisticRegression': LogisticRegression(),
    'SVC': SVC(),
    'KNeighbors': KNeighborsClassifier(),
    'GradientBoosting': GradientBoostingClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'LightGBM': LGBMClassifier()
}

# Define hyperparameter grids for each model
param_grids = {
    'RandomForest': {
        'n_estimators': [100, 200, 300],
        'max_depth': [10, 20, 30],
        'class_weight': ['balanced']
    },
    'XGBoost': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 7],
        'scale_pos_weight': [1, np.sum(y_train == 0) / np.sum(y_train == 1)]
    },
    'LogisticRegression': {
        'C': [0.1, 1, 10],
        'penalty': ['l1', 'l2'],
        'solver': ['liblinear'],
        'class_weight': ['balanced']
    },
    'SVC': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf'],
        'class_weight': ['balanced']
    },
    'KNeighbors': {
        'n_neighbors': [3, 5, 7],
        'weights': ['uniform', 'distance']
    },
    'GradientBoosting': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 7]
    },
    'AdaBoost': {
        'n_estimators': [50, 100, 150],
        'learning_rate': [0.01, 0.1, 0.2]
    },
    'LightGBM': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1],
        'max_depth': [5, 10],
        'is_unbalance': [True]
    }
}

# Set up cross-validation (StratifiedKFold to maintain class balance)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Store results for comparison
results = {}

# Loop through each model and perform GridSearchCV
for model_name, model in models.items():
    print(f"Training {model_name}...")
    
    # Set up GridSearchCV
    grid_search = GridSearchCV(estimator=model, param_grid=param_grids[model_name], 
                               cv=cv, scoring='accuracy', n_jobs=-1, verbose=1)
    
    # Fit the model
    grid_search.fit(x_train_encoded, y_train)
    
    # Get the best model and its parameters
    best_model = grid_search.best_estimator_
    best_params = grid_search.best_params_
    
    # Make predictions
    y_pred = best_model.predict(x_test_encoded)
    
    # Store the results
    results[model_name] = {
        'Best Parameters': best_params,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Classification Report': classification_report(y_test, y_pred)
    }

# Print results for each model
for model_name, result in results.items():
    print(f"\nModel: {model_name}")
    print(f"Best Parameters: {result['Best Parameters']}")
    print(f"Accuracy: {result['Accuracy']:.4f}")
    print("Classification Report:\n", result['Classification Report'])
```

```python
# Add the target (Churn) to the dataframe temporarily
df_with_target = x_train_encoded.copy()
df_with_target['churn'] = y_train

# Check correlations of numeric features with the target
correlation_matrix = df_with_target.corr()

# Print correlation of features with the target (Churn)
print("Correlation with Churn:\n", correlation_matrix['churn'].sort_values(ascending=False))
```

```python
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    return np.sqrt(chi2 / (n * (min(confusion_matrix.shape) - 1)))

# Check Cramr's V for each categorical feature
for col in cols:  # Add other categorical columns here
    cramer_v_value = cramers_v(df_with_target[col], df_with_target['churn'])
    print(f"Cramr's V for {col}: {cramer_v_value}")
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

# Instantiate the model (RandomForestClassifier)
model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)

# Define RFE for feature selection, selecting the top 20 features (adjust the number as needed)
rfe = RFE(model, n_features_to_select=20)

# Fit RFE on the training data
rfe.fit(x_train_encoded, y_train)

# Get the selected feature names
selected_features = x_train_encoded.columns[rfe.support_]
print("Selected Features:\n", selected_features)

# Reduce the training and test sets to include only the selected features
x_train_selected = x_train_encoded[selected_features]
x_test_selected = x_test_encoded[selected_features]
```

```python
from sklearn.linear_model import LogisticRegression

# Apply L2 (Ridge) regularization to Logistic Regression
logistic_model = LogisticRegression(C=0.01, penalty='l2', solver='liblinear', class_weight='balanced', random_state=42)

# Train the model with the selected features
logistic_model.fit(x_train_selected, y_train)

# Evaluate the model
y_pred_logistic = logistic_model.predict(x_test_selected)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_logistic))
```

```python
from xgboost import XGBClassifier

# Apply L1 (alpha) and L2 (lambda) regularization in XGBoost
xgb_model = XGBClassifier(n_estimators=100, learning_rate=0.01, max_depth=3, scale_pos_weight=1, 
                          alpha=0.5, reg_lambda=1.0, random_state=42)

# Train the model with the selected features
xgb_model.fit(x_train_selected, y_train)

# Evaluate the model
y_pred_xgb = xgb_model.predict(x_test_selected)
print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
```

```python
from sklearn.metrics import classification_report, accuracy_score

# Logistic Regression Evaluation
print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_logistic))

# XGBoost Evaluation
print("XGBoost Classification Report:\n", classification_report(y_test, y_pred_xgb))
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

# Define the SVC model
svc_model = SVC(random_state=42)

# Define the parameter grid
param_grid_svc = {
    'C': [0.01, 0.1, 1, 10],  # Regularization strength
    'kernel': ['rbf', 'linear'],  # RBF and linear kernel
    'gamma': ['scale', 'auto'],  # Gamma values
    'class_weight': ['balanced']  # Handling class imbalance
}

# Setup GridSearchCV for SVC
grid_search_svc = GridSearchCV(estimator=svc_model, param_grid=param_grid_svc, 
                               cv=5, scoring='accuracy', n_jobs=-1, verbose=1)

# Fit the model
grid_search_svc.fit(x_train_selected, y_train)

# Get the best parameters and results
best_svc_model = grid_search_svc.best_estimator_
best_params_svc = grid_search_svc.best_params_
print("Best parameters for SVC:", best_params_svc)

# Evaluate the model with the best parameters
y_pred_svc = best_svc_model.predict(x_test_selected)
print("SVC Accuracy:", accuracy_score(y_test, y_pred_svc))
print("SVC Classification Report:\n", classification_report(y_test, y_pred_svc))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Define the KNeighbors model
knn_model = KNeighborsClassifier()

# Define the parameter grid
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9],  # Number of neighbors
    'weights': ['uniform', 'distance'],  # Weight function used in prediction
    'p': [1, 2]  # Distance metric (1 for Manhattan, 2 for Euclidean)
}

# Setup GridSearchCV for KNeighbors
grid_search_knn = GridSearchCV(estimator=knn_model, param_grid=param_grid_knn, 
                               cv=5, scoring='accuracy', n_jobs=-1, verbose=1)

# Fit the model
grid_search_knn.fit(x_train_selected, y_train)

# Get the best parameters and results
best_knn_model = grid_search_knn.best_estimator_
best_params_knn = grid_search_knn.best_params_
print("Best parameters for KNeighbors:", best_params_knn)

# Evaluate the model with the best parameters
y_pred_knn = best_knn_model.predict(x_test_selected)
print("KNeighbors Accuracy:", accuracy_score(y_test, y_pred_knn))
print("KNeighbors Classification Report:\n", classification_report(y_test, y_pred_knn))
```
